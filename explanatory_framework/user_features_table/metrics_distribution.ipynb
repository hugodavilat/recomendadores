## Metrics Study

Notebook usado para entender as métricas e ver a distribuição delas.

```python
METRICS = ["MAE", "RMSE", "AUC", "MAP", "NDCG@100", "Precision@100", "Recall@100"]
ALGORITHMS = ["UserKNN-Amplified", "UserKNN-BM25", "UserKNN-Cosine", "UserKNN-IDF",
              "ItemKNN-AdjustedCosine", "BPR", "MF", "SVD", "PMF", "NMF"]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# import numpy as np

def get_from_metric_csv(dataset, metric, algo):
    df = pd.read_csv(f'{dataset}/{metric}.csv', sep='\s*,\s*',
                     index_col=0, engine='python')
    metric = df[[algo]].to_numpy().flatten()
#     uid = [i for i in range(1, len(metric)+1)]
    return metric

def plot_histo(x, x_label, title, bins=15):
    plt.hist(x, density=True, bins=bins, alpha=0.5, label=x_label)
    plt.ylabel('Probability (%)')
    plt.legend(loc='upper right')
#     plt.xlabel(x_label);
    plt.title(title)
```

```python
# Normalized Discount Cumulative Gain (NDCG)

metric = "NDCG@100"
algos = ["NMF", "UserKNN-Amplified", "MF", "BPR"]
dataset = "ml-1M"

for algo in algos:
    plot_histo(get_from_metric_csv(dataset, metric, algo), algo, f'{algos}@{dataset}; metric:{metric}')
```

```python
# Mean Absolute Error (MAE)

metric = "MAE"
algos = ["NMF", "UserKNN-Amplified", "MF", "BPR"]
dataset = "ml-1M"

for algo in algos:
    plot_histo(get_from_metric_csv(dataset, metric, algo), algo, f'{algos}@{dataset}; metric:{metric}')
```

```python
# Root Mean Squared Error (RMSE)

metric = "RMSE"
algos = ["NMF", "UserKNN-Amplified", "MF", "BPR"]
dataset = "ml-1M"

for algo in algos:
    plot_histo(get_from_metric_csv(dataset, metric, algo), algo, f'{algos}@{dataset}; metric:{metric}')
```

```python
# Area Under the Curve (AUC

metric = "AUC"
algos = ["NMF", "UserKNN-Amplified", "MF", "BPR"]
dataset = "ml-1M"

for algo in algos:
    plot_histo(get_from_metric_csv(dataset, metric, algo), algo, f'{algos}@{dataset}; metric:{metric}')
```

```python
# Mean Average Precision (MAP)

metric = "MAP"
algos = ["NMF", "UserKNN-Amplified", "MF", "BPR"]
dataset = "ml-1M"

for algo in algos:
    plot_histo(get_from_metric_csv(dataset, metric, algo), algo, f'{algos}@{dataset}; metric:{metric}')
```
